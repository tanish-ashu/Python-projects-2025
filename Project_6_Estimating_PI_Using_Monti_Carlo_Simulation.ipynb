**Estimating PI Using Monti Carlo Simulation**

The Monte Carlo method is a statistical Simulation technique used  to estimate mathematical values through random sampling.

To eastimate PI
we inscribe a cricle within a square.



The Probability of a random placed point falling inside the circle versus the square is proportional to their area:

**To implement this estimation in Python, Following the steps:**

importing necessary libraries.

```python
import random , matplotlib
```

Defining function that generates num times Random points in 2d space.

```python
def generatePoint():
  x = random.random()
  y = random.random()
  return x,y
```

Defining function that validate wheather the points lie with circle or not.

```python
def validate(x,y):
  if  ((x**2) + (y**2)) <= 1:
    return 1
  else:
    return 0
```

Now, Monti Carlo Estimation of PI using 1000 random points

Screenshot 2025-04-20 203144.png

```python
Points_Square = (int(input("Enter the number of points the generate (area of square)")))
Points_Circle = 0
for i in range(0, Points_Square):
 x,y = generatePoint()
 Points_Circle += validate(x,y)
print(Points_Circle)
```

```python
def valuePI( Points_Square, Points_Circle):
  return ( 4 * Points_Circle)/ Points_Square
```

```python
print(f"The value of PI obtained is: {valuePI(Points_Square,Points_Circle) }")
```
